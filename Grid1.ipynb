```python
#Import libraries numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Make 2-nd order surface function
def f(x,y):
    return np.sin(x)
```

```python
#Make plot function f 
def plot_f(x,y,z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x,y,z)
    plt.show()

#Make grid
x = np.linspace(-np.pi, np.pi, 15)
y = np.linspace(-np.pi, np.pi, 15)
x, y = np.meshgrid(x, y)
 
#Make z
z = f(x,y)

#Plot
plot_f(x,y,z)
```

```python
#Make gradient function of f
def grad_f(x,y):
    return np.array([np.cos(x), 0])

#Make regular grid
x = np.linspace(-np.pi, np.pi, 15)
y = np.linspace(-np.pi, np.pi, 15)
grid = np.array([[xx,yy] for xx in x for yy in y])
grad = np.array([grad_f(xx,yy) for xx in x for yy in y])
new_grid = grid + grad

#Conect points of grid with edges
edges = []
for i in range(0, len(grid)):
    for j in range(i+1, len(grid)):
        if np.linalg.norm(grid[i] - grid[j]) < 0.45:
            edges.append([i,j])

#Plot regular grid on 2d plot as points with size of point 5
plt.scatter(new_grid[:,0], new_grid[:,1], s=8, c='b')
#Plot edges
for edge in edges:
    plt.plot(new_grid[edge,0], new_grid[edge,1], c='b')
# plt.scatter(grid[:,0], grid[:,1], s=8, c='r')
#Plot heatmap of function f
plt.imshow(z, extent=[-np.pi,np.pi,-np.pi,np.pi], origin='lower')
plt.show()

# #Convert to 1-d array
# new_x = new_x.flatten()
# new_y = new_y.flatten()

# #Make 3d scatter plot of the new mesh grid and plot of function f on the same plot using plotly
# import plotly.graph_objects as go
# fig = go.Figure(data=[go.Scatter3d(x=new_x, y=new_y, z=f(new_x,new_y), mode='markers'),
#                         go.Surface(x=x, y=y, z=z)])
# fig.show()
```
